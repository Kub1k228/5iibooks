```python
class Rational:
    @staticmethod
    def gcd(a,b):
        while (b != 0):
            (a,b) = (b,a%b)
        return a
    @staticmethod
    def sgn(x):
        if x>0:
            return 1
        elif x<0:
            return -1
        else:
            return 0
    def __init__(self,n,d):
        if n==0:
            self.num=0
            self.den=1
        else:
            z=self.sgn(n)*self.sgn(d)
            n=abs(n)
            d=abs(d)
            k=self.gcd(n,d)
            self.num=z*n//k
            self.den=d//k
    def __str__(self):
        if self.num==0:
            return "0"
        else:
            return str(self.num)+"/"+str(self.den)
    def __add__ (self,o):
        n1=self.num
        d1=self.den
        if type(o)==int:
            n2=o
            d2=1
        else:
            n2=o.num
            d2=o.den
        n=n1*d2+n2*d1
        d=d1*d2
        return Rational(n,d)
    def __add__(self,o):
        n1=self.num
        d1=self.den
        if type(o)==int:
            n2=0
            d2=1
        else:
            n2=o.num
            d2=o.den
        n=n1*d2+n2*d1
        d=d1*d2
        return Rational(n,d)
    def __sub__(self,o):
        n1=self.num
        d1=self.den
        n2=o.num
        d2=o.den
        n=n1*d2-n2*d1
        d=d1*d2
        return Rational(n,d)
    def __mul__(self,o):
        n1=self.num
        d1=self.den
        n2=o.num
        d2=o.den
        n=n1*n2
        d=d1*d2
        return Rational(n,d)
    def __floordiv__(self,o):
        n1=self.num
        d1=self.den
        n2=o.num
        d2=o.den
        n=n1*d2
        d=d1*n2
        return Rational(n,d)
    

d1=Rational(1,2)
d2=Rational(1,3)
d3=d1+d2
print(d3)
d4=d1-d2
print(d4)
d5=d1*d2
print(d5)
d6=d1*d2
print(d6)
d7=d1//d2
print(d7)
d8=d1+6
print(d8)
```

```python
import numpy as np
class Trigonometry:
    def __init__(self,param1) -> None:
        self.p = param1
    def sin(self):
        return np.sin(self.p)
    def cos(self):
        return np.cos(self.p)
    def tan(self):
        return np.tan(self.p)
    def arccos(self):
        return np.arccos(self.p)
    def arcsin(self):
        return np.arcsin(self.p)
    def arcatan(self):
        return np.arctan(self.p)
    def to_rad(self):
        return np.radians(self.p)
    
t1 = Trigonometry (1.0)
print(t1.sin())
print(t1.cos())
print(t1.tan())
print(t1.arcsin())
print(t1.arccos())
print(t1.arcatan())
print(t1.to_rad())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/AnnaShestova/salary-years-simple-linear-regression/master/Salary_Data.csv"
dataset = pd.read_csv(url)
dataset.head()
```

```python
print (dataset.shape)
dataset.describe()
```

```python
plt.scatter(dataset['YearsExperience'], dataset['Salary'],color = 'b',label=' ')
plt.xlabel(" ()")
plt.ylabel(" ")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.model_selection import train_test_split
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1].values
print(X)
print(y)

X_train, X_test ,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)

```

```python
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=[30,20],dpi=300)
tree.plot_tree(regressor,fontsize=7)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
df = pd.DataFrame({"Actual":y_test,'Predicted':y_pred})
df
```

```python
from sklearn import metrics
print ('Mean Squared error',metrics.mean_squared_error(y_test,y_pred))
print ('Mean Absolute error',metrics.mean_absolute_error(y_test,y_pred))
metrics.mean_absolute_error(y_test,y_pred)/np.average(y)*100
```
