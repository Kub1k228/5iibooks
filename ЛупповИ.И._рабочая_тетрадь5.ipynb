```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/AnnaShestova/salary-years-simple-linear-regression/master/Salary_Data.csv"
dataset = pd.read_csv(url)
dataset.head()
```

```python
print (dataset.shape)
dataset.describe()
```

```python
plt.scatter(dataset['YearsExperience'], dataset['Salary'],color = 'b',label=' ')
plt.xlabel(" ()")
plt.ylabel(" ")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
from sklearn.model_selection import train_test_split
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1].values
print(X)
print(y)

X_train, X_test ,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)

```

```python
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=[30,20],dpi=300)
tree.plot_tree(regressor,fontsize=7)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
df = pd.DataFrame({"Actual":y_test,'Predicted':y_pred})
df
```

```python
from sklearn import metrics
print ('Mean Squared error',metrics.mean_squared_error(y_test,y_pred))
print ('Mean Absolute error',metrics.mean_absolute_error(y_test,y_pred))
metrics.mean_absolute_error(y_test,y_pred)/np.average(y)*100
```

```python
#
url = 'https://raw.githubusercontent.com/akmand/datasets/main/carvana_kicks_clean.csv'
dataframe = pd.read_csv(url)
dataframe.head(10)
```

```python
param1 = 'VehicleAge'
param2 = 'MMRAcquisitionRetailAveragePrice'
plt.scatter(dataframe[param1],dataframe[param2],color= 'b', label = 'exam data')
plt. xlabel(param1)
plt.ylabel("avarage price")
plt.show()
```

```python
c1 = list(dataframe[param1])
c2 = list(dataframe[param2])
dict = {param1:c1,param2:c2}
dataset = pd.DataFrame(dict) 
dataset.head(10)
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1].values

X_train, X_test ,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=0)

regressor = DecisionTreeRegressor()
regressor.fit(X_test,y_test)
```

```python
plt.figure(figsize=[30,20],dpi=300)
tree.plot_tree(regressor,fontsize=7,max_depth=40)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
df = pd.DataFrame({"Actual":y_test,'Predicted':y_pred})
df
```

```python
print ('Mean Squared error',metrics.mean_squared_error(y_test,y_pred))
print ('Mean Absolute error',metrics.mean_absolute_error(y_test,y_pred))
metrics.mean_absolute_error(y_test,y_pred)/np.average(y)*100
```
